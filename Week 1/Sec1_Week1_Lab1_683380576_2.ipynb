### Lab 1.2: Working with Variables and Basic Data Types
- Objective: Students will define variables of different data types, perform simple operations, and display results.


2. Declare and print variables:


```python
quantity = 10
price = 195.5
title = "Basic Python 101"
available = True

print("Book title : ", title, "\t" ,type(title))
print("Book available : ", available, "\t\t", type(available))
print("Book quantity : ", quantity, "\t\t", type(quantity))
print("Book price : ", price, "\t\t", type(price))
```

3. Perform simple arithmetic:


```python
num1 = 250
num2 = 5

sum = num1 + num2
print("num1 + num2 = ",sum)

sum = num1 - num2
print("num1 - num2 = ",sum)

sum = num1 * num2
print("num1 * num2 = ",sum)

sum = num1 / num2
print("num1 / num2 = ",sum)
```

4. String concatenation and f-strings:


```python
First_name = "Tanatorn"
Last_name = "Palang"
Full_name = First_name + " " + Last_name
print(f"Hello, {Full_name}")
```

5. User input and type conversion:


```python
num = input("Type your biggest favourite number in the world : ")
print("Your number is", num, type(num))
result = int(num) + 5
print(f"So your lucky number is {num} + 5 = {result}")
```
