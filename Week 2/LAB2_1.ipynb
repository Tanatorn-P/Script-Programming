```python
number = int(input("Please type the number : "))

isPositive = "positive"
isEven = "even"

if number==0:
  isPositive = "zero"
elif number<0:
  isPositive = "negative"

if number%2 != 0:
  isEven = "odd"

print(f"The number {number} is {isPositive} and {isEven}")
```
