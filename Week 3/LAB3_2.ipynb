### Lab 3.2 : Countdown and User guesing game


```python
#Part 1: Countdown Timer
while True:
  try:
    number2 = int(input("Insert your number : "))
    if (number2 > 0):
      while (number2 >= 0):
        print(number2)
        number2 -= 1
      print("Blast off!")
    else:
      print("Please insert number greater than 0")
    break
  except ValueError :
    print("\nใส่ตัวเลข บักปอบ\n")
```

```python
#Part 2: Simple Guessing Game
import random
rand_number = random.randint(1, 50)
guess = 0

while guess < 5:
  try:
    number3 = int(input("Guess a number 1 - 50 : "))

    if number3 == rand_number:
      print("Congratulations! You guessed it!")
      break

    elif number3 < rand_number:
      guess += 1
      print(f"--{number3} is too low! Try again.  ({5 - guess} times left)\n")

    elif number3 > rand_number:
      guess += 1
      print(f"--{number3} is too high! Try again. ({5 - guess} times left)\n")

  except ValueError:
    guess += 1
    print(f"--ใส่ตัวเลข บักปอบ ({5 - guess} times left)\n")

  if guess == 5:
    print("--Sorry you lose. You've guess for 5 times!")
    print(f"--The correct number is : {rand_number}")
```
