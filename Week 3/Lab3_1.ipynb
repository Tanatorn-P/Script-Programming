### Lab 3.1 : Multiple Number generater


```python
#Part 1: Single Number Table
while True:
  try:
    number = int(input("Insert your number : "))
    for i in range(1, 13):
      print(f"{number} x {i:<2} = {i*number:<}")
    break
  except ValueError:
    print("\nใส่ตัวเลข บักปอบ\n")
```

```python
#Part 2: Full Multiplication Grid
for i in range(1, 13):
  for j in range(1, 13):
    print(f"{i*j:<4}", end = "")
  print("")
```
