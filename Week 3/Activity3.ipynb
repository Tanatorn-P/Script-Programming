### Activity 3.2: Pattern Printing with Nested Loops

```python
# Square of asterisks
while True:
  try:
    number4 = int(input("Insert your asterisks size : "))

    for i in range(0, number4):
      for j in range(0, number4):
        print("*", end="")
      print()
    break
  except:
    print("--ใส่ตัวเลข บักปอบ\n")
```

```python
# Right-angled triangle of asterisks
while True:
  try:
    number4 = int(input("Insert your asterisks size : "))
    count = 0
    for i in range(0, number4):
      if count <= number4:
        for j in range(0, count+1):
          print("*", end="")
        print()
      count += 1
    break
  except:
    print("--ใส่ตัวเลข บักปอบ\n")
```
