### LAB 4.1 Student Management System (List Operations)

```python
student_names = []
con = 0

print("\n----------------------------------------------")
print("          Student Management System")
print("----------------------------------------------")
while True:
  try:
    print("1. Show student list")
    print("2. Add student")
    print("3. Delete student")
    print("4. Search student")
    print("5. Sort student")
    print("6. Exit")
    con = int(input("Enter the menu number (1-6). : "))

    if con == 1: # แสดงชื่อทั้งหมด
      print("\n----------------------------------------------")
      if len(student_names) != 0:
        print("   List of student   ")
        print("----------------------------------------------")
        for i in range(len(student_names)):
          print(f"{i+1}. {student_names[i]}")
      else:
        print("Student list is empty.")
      print("----------------------------------------------\n")

    elif con == 2: # เพิ่มชื่อ
      print("\n----------------------------------------------")
      new_student = input("Add new student : ").strip().title()

      if not new_student:
        print("Name cannot be empty!")
      elif new_student in student_names:
        print(f"'{new_student}' already exists in the system!")
      else:
        student_names.append(new_student)
        print(f"'{new_student}' added successfully!")
      print("----------------------------------------------\n")

    elif con == 3: # ลบชื่อ
      print("\n----------------------------------------------")
      if not student_names:
        print("Student list is empty!")
      else:
        remove = input("Enter student name to remove : ").strip().title()
        if not remove:
          print("Name cannot be empty!")
        elif remove not in student_names:
          print(f"'{remove}' is not found in system!")
        else:
          student_names.remove(remove)
          print(f"'{remove}' deleted successfully!")
      print("----------------------------------------------\n")

    elif con == 4: # เสิร์ชชื่อ
      print("\n----------------------------------------------")
      if not student_names:
        print("Student list is empty!")
      else:
        search = input("Search student name : ").strip().title()

        if not search:
          print("Name cannot be empty!")
        elif search not in student_names:
          print(f"'{search}' is not found in system!")
        else:
          print(f"'{search}' is in the No.{student_names.index(search)+1} of student list.")
        print("----------------------------------------------\n")

    elif con == 5: # เรียงลำดับชื่อ
      print("\n----------------------------------------------")
      if not student_names:
        print("Student list is empty!")
      else:
        student_names.sort()
        print("Student list after sorting : ")
        for i in range(len(student_names)):
          print(f"{i+1}. {student_names[i]}")
      print("----------------------------------------------\n")

    elif con == 6: # จบโปรแกรม
      print("\n----------------------------------------------\n")
      print("THANK YOU...")
      break
  except ValueError:
    print("\n----------------------------------------------")
    print("Please type number of the menu (1-6).")
    print("----------------------------------------------\n")
```

### LAB 4.2 Geometric Calculations (Tuples & Immutability)

```python
# lab4_2.py - Geometric Calculations (Tuples & Immutability)
# หมายเหตุ: โค้ดนี้ตัดส่วน "Challenge" (รับพิกัดจากผู้ใช้เพื่อคำนวณระยะทาง) ออก
#          เพื่อให้นักศึกษานำ concept ด้านล่างไปต่อยอดคิดเองเป็นแบบฝึกหัด

import math

print("--- Lab 4.2: Geometric Calculations ---")

# 1) กำหนดจุดสองจุดด้วย tuple: point1 และ point2 *** เพิ่ม challenge แล้ว ***
point1 = tuple(map(int, input("Enter the Coordinates for point 1 (x1, y1) : ").split(","))) #เติมพิกัด x y ของจุด 👈ต้องทำ
print(point1)
point2 = tuple(map(int, input("Enter the Coordinates for point 2 (x2, y2) : ").split(",")))  #เติมพิกัด x y ของจุด 👈ต้องทำ
print(point2)

# 2) เข้าถึงค่าพิกัด x, y ของแต่ละจุด
print("\n[1] Access Coordinates")
print(f"Point 1: x = {point1[0]}, y = {point1[1]}") #เติมเลข 0 และ 1 ใน [] 👈ต้องทำ
print(f"Point 2: x = {point2[0]}, y = {point2[1]}") #เติมเลข 0 และ 1 ใน [] 👈ต้องทำ

# 3) ทดลองแก้ไขค่าใน tuple เพื่อแสดงให้เห็นว่า tuple เปลี่ยนแปลงไม่ได้ (immutable)
print("\n[2] Attempt Modification (Immutability Demonstration)")
try:
    point1[0] = 5  # บรรทัดนี้จะทำให้เกิด TypeError เพราะ tuple แก้ไขไม่ได้
except TypeError as e:
    print(f"Error trying to modify tuple: {e}")

# 4) คำนวณระยะห่างระหว่างจุดสองจุดด้วยสูตรระยะทาง
#    distance = sqrt( (x2 - x1)^2 + (y2 - y1)^2 )
print("\n[3] Calculate Distance")
distance = math.sqrt((point2[0] - point1[0]) ** 2 + (point2[1] - point1[1]) ** 2) #เติมเลขตำแหน่งในลิส 0 และ 1 ใน [] ให้ถูกต้อง 👈ต้องทำ
print(f"Distance between point1 {point1} and point2 {point2}: {distance}")

# --------------------------------------------------------------------
# ให้นักศึกษาลองต่อยอดเอง (Challenge):
# ลองแก้โค้ดให้รับพิกัดของจุดใหม่ 2 จุดจากผู้ใช้ผ่าน input()
# (อย่าลืมแปลงชนิดข้อมูลจาก input() ให้เป็นตัวเลขก่อนนำไปคำนวณ)
# แล้วคำนวณระยะห่างระหว่างจุดทั้งสองที่ผู้ใช้กรอกเข้ามา
# --------------------------------------------------------------------
```

### Activity 4.1 List vs. Tuple: Scenario Matching

**สถานการณ์ต่อไปนี้ แต่ละเหตุการณ์ควรใช้ list หรือ tuple เพราะอะไร**<br>
1. ข้อมูลอุณหภูมิประจำวันที่มีการบันทึกไว้ตลอดหนึ่งสัปดาห์<br>
ตอบ List เพราะอุณหภูมิมีการบันทึกเพิ่มใหม่ทุกวัน<br><br>
2. ค่าสี RGB ของเฉดสีเฉพาะเฉดหนึ่ง (เช่น สีแดงคือ (255, 0, 0))<br>
ตอบ Tuple เพราะหากมีการเปลี่ยนค่าสีจะเปลี่ยนไป<br><br>
3. รายการสินค้าในรถเข็นชอปปิงที่สามารถกดเพิ่มหรือลบออกได้<br>
ตอบ List เพราะมีการแก้ไขรายการสินค้า<br><br>
4. รายชื่อวันในหนึ่งสัปดาห์เรียงตามลำดับ (จันทร์ - อาทิตย์)<br>
ตอบ Tuple เพราะลำดับของวันไม่ควรถูกเปลี่ยนแปลง<br><br>
5. คะแนนเก็บของนักเรียนจากการส่งงานชุดต่าง ๆ ซึ่งอาจมีการอัปเดตหรือแก้ไขคะแนนในภายหลังได้<br>
ตอบ List เพราะมีการแก้ไขคะแนน<br><br>
6. พิกัด GPS ของสถานที่สำคัญหรือจุดแลนด์มาร์กที่ตั้งอยู่กับที่<br>
ตอบ Tuple เพราะพิกัดของแผนที่ต้องคงเดิมเสมอ

### Activity 4.2 Interactive To-Do List (using List Operations)

```python
todo_list = ["A", "B", "C"]
comp_list = []
cont = 0

print("----------------------------")
print("        To-Do List")
print("----------------------------")

while True:
  try:
    print("1. Add a task")
    print("2. View task")
    print("3. Mark task as complete")
    print("4. Exit")
    cont = int(input("Enter number of the menu (1-4) : "))

    if cont == 1:
      print("\n----------------------------")
      new_task = input("Enter your new task : ").strip().title()

      if not new_task:
        print("Task can not be empty!")
      elif new_task in todo_list:
        print("Task already in To-do list!")
      else:
        todo_list.append(new_task)
        print("Task added successfully!")
      print("----------------------------\n")

    elif cont == 2:
      print("\n----------------------------")
      if not todo_list:
        print("To-do list is empty!")
      else:
        print("All tasks in to-do list :")
        for i, task in enumerate(todo_list):
          print(f"{i+1}: {task}")
      print("----------------------------\n")

    elif cont == 3:
      print("\n----------------------------")
      if not todo_list:
        print("To-do list is empty!")
      else:
        try:
          comp_task = int(input("Enter the number of task in the list to complete : ").strip()) -1

          if comp_task == "":
            print("Number of task can not be empty!")
          elif comp_task > len(todo_list) or comp_task < 0:
            print(f"Please Enter an task number! (1-{len(todo_list)})")
          else:
            for i, task in enumerate(todo_list):
              if i == comp_task:
                comp_list.append(todo_list.pop(i))
                print(f"Task {i+1}. had marked as complete!")
        except ValueError:
          print("Please enter number of a task")
      print("----------------------------\n")

    elif cont == 4:
      print("\n\n----------------------------")
      print("THANK YOU...")
      break

  except ValueError:
    print("Please enter number of a Menu")
```
